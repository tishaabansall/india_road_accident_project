```python
import pandas as pd
```

```python
df = pd.read_csv('../data/india_road_accidents.csv')
```

```python
print(df.columns)
print(df.shape)
```

```python
df = df.drop_duplicates()
```

```python
df = df.dropna(subset=[
    'State Name',
    'City Name',
    'Accident Severity',
    'Time of Day'
])
```

```python
df['Hour'] = pd.to_datetime(df['Time of Day'], errors='coerce').dt.hour
```

```python
print("Final shape:", df.shape)
df.head()
```

```python
df.to_csv('../data/cleaned_accidents.csv', index=False)
```
