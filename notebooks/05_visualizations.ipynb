```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

if not os.path.exists('../plots'):
    os.makedirs('../plots')

# Load the cleaned ML dataset
df = pd.read_csv('../data/final_accidents_ml.csv')
```

```python
gender_severity = (
    df.groupby('Driver Gender', dropna=False)['Severity_Encoded']
    .mean()
    .reset_index()
)
plt.figure(figsize=(5,4))
sns.barplot(data=gender_severity, x='Driver Gender', y='Severity_Encoded')
plt.title('Average Accident Severity by Driver Gender')
plt.ylabel('Avg Severity (0=Minor, 2=Fatal)')
plt.savefig('../plots/gender_vs_severity.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Accident Severity', y='Speed Limit (km/h)', data=df)
plt.title('Speed Limit vs Accident Severity')
plt.savefig('../plots/speed_vs_severity.png', bbox_inches='tight')
plt.show()
```

```python
state_counts = df['State Name'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=state_counts.values, y=state_counts.index)
plt.title('Top 10 States by Accident Count')
plt.xlabel('Accident Count')
plt.ylabel('State')
plt.savefig('../plots/accidents_by_state.png', bbox_inches='tight')
plt.show()
```

```python
day_night_df = pd.DataFrame({
    'Time': ['Day', 'Night'],
    'Count': [
        day_night_counts.get(0, 0),
        day_night_counts.get(1, 0)
    ]
})

plt.figure(figsize=(5,4))
sns.barplot(data=day_night_df, x='Time', y='Count')
plt.title('Accidents: Day vs Night')
plt.ylabel('Accident Count')
plt.savefig('../plots/day_vs_night.png', bbox_inches='tight')
plt.show()

```
